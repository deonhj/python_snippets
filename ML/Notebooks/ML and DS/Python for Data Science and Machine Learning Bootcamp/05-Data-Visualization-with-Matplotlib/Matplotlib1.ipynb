```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

### **FUNCTIONAL METHOD**

```python
plt.plot(x, y, )
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
# use plt.show() to show the plot in an environment outside of jupyter
```

## plt.subplot(rows, columns, plot_number)

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g--')
```

### **Object oriented method of using Matplotlib**

add_axes([left_of_axes, bottom_of_axes, width, height])

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y, 'b--')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

add_axes([left_of_axes, bottom_of_axes, width, height])

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')

axes1.plot(x, y, 'b--')
axes2.plot(y, x, 'r--')
```

## **Subplots with Object Oriented Method**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
axes[0][1].plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)
plt.tight_layout()

axes[0].set_title('FIRST')
axes[1].set_title('SECOND')
axes[-1].set_title('LAST')

axes[0].plot(x, y)
axes[-1].plot(y, x)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
axes[2][1].plot(x, y)
```

## **Figure Size and DPI**

```python
fig = plt.figure(figsize=(6, 3))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

## **Config for subplots**

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 4))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig('Latest.jpg', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label='x squared')
ax.plot(x, x ** 3, label='x cubed')
ax.legend(loc='best')
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x ** 2, label='x squared')
ax.plot(x, x ** 3, label='x cubed')
ax.legend(loc=(0.1, 0.2))
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#3275a8') # RGB Hrx or simple names like 'blue'
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#3275a8', linewidth=5, linestyle='-', 
        alpha=0.9, marker='^', markersize=20, markerfacecolor='red', 
        markeredgewidth=4, markeredgecolor='#000000') # RGB Hrx or simple names like 'blue'
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#3275a8', lw=5, ls='--')
ax.set_xlim([2,3])
ax.set_ylim([5, 20])
```
