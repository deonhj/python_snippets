```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
sns.barplot(x='smoker',y='total_bill',data=tips, estimator=sum)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='smoker', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python

sns.stripplot(x='day',y='total_bill',data=tips, hue='smoker', dodge=True, size=2)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips, hue='sex', dodge=True, size=3)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', dodge=True, size=2)
sns.swarmplot(x='day',y='total_bill',data=tips, hue='sex', dodge=True, size=3, color='black')
```

```python
sns.catplot(x='day',y='total_bill',data=tips, kind='bar')
```
