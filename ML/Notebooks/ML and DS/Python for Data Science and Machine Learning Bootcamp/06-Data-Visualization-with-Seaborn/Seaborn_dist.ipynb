```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
sns.set_style('darkgrid')
sns.displot(tips['total_bill'], kde=True)
```

```python
sns.displot(tips['total_bill'], kde=True, bins=15)
```

```python
graph_v = sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', color='#ba3030')
graph_v.set_axis_labels('Tips', 'Total Bill')
```

```python
sns.pairplot(tips, kind='reg', hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```
